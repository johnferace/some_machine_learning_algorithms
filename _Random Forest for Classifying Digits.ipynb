```python
#Utilizamos la undataset de la libreria sklearn llamado load_digits que contienen imagenes de numeros 
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.keys()
```

```python
#Visualizamos algunos datos del dataset
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(6, 6)) # figure size in inches
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)
# plot the digits: each image is 8x8 pixels
for i in range(64):
 ax = fig.add_subplot(8, 8, i + 1, xticks=[], yticks=[])
 ax.imshow(digits.images[i], cmap=plt.cm.binary, interpolation='nearest')
 # label the image with the target value
 ax.text(0, 7, str(digits.target[i]))
```

```python
#CClasificamos los digitos usando un algoritmo de clasificaion llamado bosques aleatorios
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
Xtrain, Xtest, ytrain, ytest = train_test_split(digits.data, digits.target,random_state=0)
model = RandomForestClassifier(n_estimators=1000)
model.fit(Xtrain, ytrain)
ypred = model.predict(Xtest)
```

```python
#vemos algo de imformacion del entrenamiento
```

```python
from sklearn import metrics
print(metrics.classification_report(ypred, ytest))
```

```python
#Por ultimo utilizamos una matriz de confusion para ver los resultados 
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(ytest, ypred)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('true label')
plt.ylabel('predicted label');
```
